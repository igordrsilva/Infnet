```python
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
```

## üìè Regress√£o

```python
train_f, test_f, train_t, test_t = train_test_split(digits.data, digits.target, test_size=10, stratify=digits.target)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn = knn.fit(train_f, train_t)

predicted = knn.predict(test_f)
```

```python
plt.figure(figsize=(6,6))

plt.scatter(test_t, predicted, alpha=0.7)

plt.xlabel('Valores Reais')
plt.ylabel('Valores Previstos')
plt.title('Real vs Previsto')
plt.show()
```
